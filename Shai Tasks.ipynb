### Task 3 Data Cleaning: 
Handle missing data by suitable method with explain why you use it.

##### A) Dropping the null columns

```python
# Droping columns that have no values
df.drop(["Notes", "Status"], axis = 1, inplace = True)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# Droping rows that doesn't have a valied balues
df.drop(df[df["EmployeeName"] == "Not provided"].index, inplace= True)
```

```python
# Make the job title have same letters case 
df["JobTitle"] = df["JobTitle"].apply(lambda x: x.title())
```

##### B) Imputation the missing base pay using the average base pay for each job title

```python
# Creating a List of job titles that have a base pay as a missing value
MissingBasePay = df[df["BasePay"].isnull()]["JobTitle"].unique().tolist()

# Grouping jobs based on the mean base pay 
basepay_grouped = df.groupby("JobTitle").mean(numeric_only= True)[["BasePay"]]

basepay_grouped[:5]
```

```python
# Creating a dictionary with the job title and its mean salary
basepay_dict = basepay_grouped.to_dict()
basepay_dict = basepay_dict["BasePay"]
```

```python
# Creating a dictionary that contain the mean salary for missing job titles
basepay_imputation = {}
for i in MissingBasePay:
  if i in basepay_imputation:
    continue
  elif i in basepay_dict:
      basepay_imputation[i] = basepay_dict.get(i)
  else:
    basepay_imputation[i] = "Not Found"
```

```python
# Filling the missing value
df["BasePay"] = df["BasePay"].fillna(df["JobTitle"].map(basepay_imputation))
```

```python
df.isnull().sum()
```

```python
df[df["BasePay"].isna()]
```

```python
# Checking the benefits columns
df[df["Benefits"].isna()][:5]
```

```python
'''Since TotalPayBenefits is the sum of Benefits and TotalPay 
we can see the NaN valeus can be repleaced with zero it also make since that benefits can take a value of 0
the total pay is the sum of base pay, overtmie and the other pay 
so the remanining missing values can be replaced with zero as well'''

df.fillna(0,inplace = True)
```

#### C) Creating the department table

```python
# Creating a copy of the data frame
df_copy = df.copy()
```

```python
pattern = r'\((.*?)\)'

df = df.merge(df["JobTitle"].str.extract(pattern),left_index=True, right_index = True).rename(columns={0:"Department"})
df.head(3)
```

```python
# Creating Other categories for employees without a department
df['Department'].replace(np.nan, "Other/unspecified", inplace=True)

# Remove the depratment from the employee name
df["JobTitle"].replace(pattern,"",regex= True, inplace= True)

df[["JobTitle", "Department"]].head()
```

###### F) Creating new feature that specify the period of how long the company stayed in the firm

```python
# Creating new feature which determine the period for each employee in the firm

max_date = df["Year"].max()
df["YearsInCompany"] = df["Year"].apply(lambda x : max_date - x)

df.head()
```
